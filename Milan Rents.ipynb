**2. What is the most unbiased and efficient way to compute the best predictive algorithm?**

To determine the most unbiased and efficient predictive algorithm, I implemented a two-step approach. First, I compared various models, including all 54 regressor models in scikit-learn, 8 custom PyTorch neural networks, XGBoost, and LightGBM, using 5-fold cross-validation on a dataset with 13,188 observations and approximately 90 features (the number of teatures changes depending on the train set). Each cross-validation fold was also independently preprocessed to prevent data leakage. Cross validations of the same model are fitted sequentially in order to implement an early stopping mechanism that terminates underperforming models, conserving computational resources.

The `HT_with_nested_CV` function fine-tunes hyperparameters using nested cross-validation. This method uses two layers of cross-validation: an outer loop to assess model performance and an inner loop to tune hyperparameters, guaranteering an unbiased selection of the optimal predictive model.

In the next cells I convert time in year and month in a way that a ML algorithm can interpret them cyclically.

Often datetime data are just divided into month and year, and then just encoded categorically or as dummies.
The problem with this is that I need to let the algorithm know that features such as these occur in cycles.
If I do not do that, it will interpret December (month 12) as being very far away from January (month 1).

One solution is to maps each cyclical variable onto a circle, creating only 2 variables.
This method is the sine and cosine transformation.
For more info: https://blog.davidkaleko.com/feature-engineering-cyclical-features.html

I will therefore encode:
- Months with a sine and cosine transformation
- Years with dummies (having only 3 years, any detectable cycle is unlikely to be captured by years, 
and differently from days and months, the first year is not close to the last one)

### Cleaning the Dataset - Preprocessing function for CV

I collect all the cleaning in a single function, that will be applied independently to every training set during cross-validation.

```python

def preprocessing_function(X, y):

    X = pd.get_dummies(X, columns=['condition', 'energy_class'], prefix=None, drop_first=True, dummy_na=True)
    X = X.astype({col: int for col in X.select_dtypes(include=['uint8', 'bool']).columns})
    X['datetime'] = pd.to_datetime(X['datetime'])
    X['year'] = X['datetime'].dt.year.astype('category')
    X['month'] = X['datetime'].dt.month.astype('category').cat.codes
    X.drop('datetime', axis=1, inplace=True)
    X['month_sin'] = np.sin(2 * np.pi * X["month"]/12)
    X['month_cos'] = np.cos(2 * np.pi * X["month"]/12)
    X.drop(['month'], axis=1, inplace=True)
    X = pd.get_dummies(X, columns=['year'], prefix=None, drop_first=True, dummy_na=False)
    X.drop(['sea_view'], axis=1, inplace=True)
    X = pd.get_dummies(X, columns=['central_heating'], prefix=None, drop_first=True, dummy_na=True)
    X = X.astype({col: int for col in X.select_dtypes(include=['uint8', 'bool']).columns})
    X = X.dropna()
    non_na_indices = X.index
    y = y.loc[non_na_indices]
    district_counts = X['district'].value_counts()
    district_to_replace = district_counts[district_counts < len(y)/100].index
    X.loc[X['district'].isin(district_to_replace), 'district'] = 'Other'
    X = pd.get_dummies(X, columns=['district'], prefix=None, drop_first=False, dummy_na=True)
    X.drop(['tv_system','alarm_system','doorman','full_ownership'], axis=1, inplace=True)
    
    return X, y
```

Manually construting Neural Network Estimators.
The only Neural Network included in SkLearn is a very basic regressor (MLPRegressor) that often performs poorly and fits slowly compared to Torch Neural Networks.
Instead, the Neural Networks constructed will be:
- 2HL
- 3HL
- 4HL
- 4HL that uses flat layers
- 5HL that uses less neurons for each layers
- 3HL with Early Stopping
- 7HL with 95% PCA
- 8HL with 90% PCA

(HL: Hidden Layers)

The following list has been constructed to compare as many differen situations as possible while not impacting drastically on computational time. If Neural Networks will happen to perform better than other methods, new Networks will be constructed and compared during hyperparameter tuning.

```python

# Defining main Neural Netowrk class

class NN(BaseEstimator, RegressorMixin):
    def __init__(self, hidden_layer_sizes=(200, 100), epochs=100, batch_size=32, activation=nn.ReLU):
        """
        Initializes the neural network regressor with specified hidden layers, epochs, batch size, and activation function.
        
        Parameters:
        hidden_layer_sizes (tuple): Sizes of hidden layers.
        epochs (int): Number of epochs for training.
        batch_size (int): Size of each training batch.
        activation (class): Activation function to use.
        """

        self.hidden_layer_sizes = hidden_layer_sizes
        self.epochs = epochs
        self.batch_size = batch_size
        self.activation = activation
        self.model_ = None
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    def build_model(self, input_dim):
        """
        Builds the neural network model with the given input dimension.
        
        Parameters:
        input_dim (int): Dimension of the input features.
        
        Returns:
        nn.Sequential: Constructed neural network model.
        """
        
        layers = []
        for size in self.hidden_layer_sizes:
            layers.append(nn.Linear(input_dim, size))
            layers.append(self.activation())
            input_dim = size
        layers.append(nn.Linear(input_dim, 1))  # Output layer
        model = nn.Sequential(*layers)
        return model

    def fit(self, X, y):
        """
        Trains the neural network on the given data.
        
        Parameters:
        X (array-like): Training data features.
        y (array-like): Training data targets.
        
        Returns:
        self: Fitted estimator.
        """
        
        # Convert input data to torch tensors and move to the appropriate device
        X = torch.tensor(X, dtype=torch.float32).to(self.device)
        y = torch.tensor(y, dtype=torch.float32).view(-1, 1).to(self.device)
        
        # Build the model
        input_dim = X.shape[1]
        self.model_ = self.build_model(input_dim).to(self.device)
        
        # Define the loss function and optimizer
        criterion = nn.MSELoss()
        optimizer = optim.Adam(self.model_.parameters())
        
        # Training loop
        for epoch in range(self.epochs):
            self.model_.train()
            optimizer.zero_grad()
            outputs = self.model_(X)
            loss = criterion(outputs, y)
            loss.backward()
            optimizer.step()
        return self

    def predict(self, X):
        """
        Predicts the target values for the given input data.
        
        Parameters:
        X (array-like): Input data features.
        
        Returns:
        array: Predicted target values.
        """
        X = torch.tensor(X, dtype=torch.float32).to(self.device)
        with torch.no_grad():
            predictions = self.model_(X).cpu().numpy().flatten()
        return predictions

    def score(self, X, y):
        """
        Returns the coefficient of determination R^2 of the prediction.
        
        Parameters:
        X (array-like): Test data features.
        y (array-like): True target values.
        
        Returns:
        float: R^2 score.
        """
        from sklearn.metrics import r2_score
        return r2_score(y, self.predict(X))
```

```python
# Early Stopping subclass

# Define a class NN_Early_Stop that inherits from the NN class
class NN_Early_Stop(NN):
    # Initialize the class with specific parameters, including early stopping parameters
    def __init__(self, hidden_layer_sizes=(200, 100, 50, 10), epochs=100, batch_size=32, activation=nn.ReLU, patience=10, validation_fraction=0.1):
        super().__init__(hidden_layer_sizes, epochs, batch_size, activation)
        self.patience = patience  # Number of epochs with no improvement after which training will be stopped
        self.validation_fraction = validation_fraction

    def fit(self, X, y):
        X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=self.validation_fraction) # Split the data into training and validation sets

         # Convert the data to PyTorch tensors and move them to the appropriate device (CPU or GPU)
        X_train = torch.tensor(X_train, dtype=torch.float32).to(self.device)
        y_train = torch.tensor(y_train, dtype=torch.float32).view(-1, 1).to(self.device)
        X_val = torch.tensor(X_val, dtype=torch.float32).to(self.device)
        y_val = torch.tensor(y_val, dtype=torch.float32).view(-1, 1).to(self.device)
        
        input_dim = X_train.shape[1] # Get the input dimension for building the model
        
        self.model_ = self.build_model(input_dim).to(self.device) # Define the loss function and the optimizer
        criterion = nn.MSELoss()
        optimizer = optim.Adam(self.model_.parameters())

        best_loss = np.inf # Initialize the best loss to infinity
        epochs_no_improve = 0 # Counter for epochs with no improvement

        # Training loop
        for epoch in range(self.epochs):
            self.model_.train()  # Set the model to training mode
            optimizer.zero_grad()  # Zero the gradients
            outputs = self.model_(X_train)  # Forward pass
            loss = criterion(outputs, y_train)  # Compute the loss
            loss.backward()  # Backward pass
            optimizer.step()  # Update the weights

            self.model_.eval()  # Set the model to evaluation mode
            with torch.no_grad():  # Disable gradient computation
                val_outputs = self.model_(X_val)  # Forward pass on validation data
                val_loss = criterion(val_outputs, y_val)  # Compute the validation loss
            
            # Check if the validation loss has improved
            if val_loss < best_loss:
                best_loss = val_loss  # Update the best loss
                epochs_no_improve = 0  # Reset the counter for epochs with no improvement
            else:
                epochs_no_improve += 1  # Increment the counter for epochs with no improvement

            # If the number of epochs with no improvement reaches the patience threshold, stop training
            if epochs_no_improve == self.patience:
                break
                
        return self  # Return the trained model instance                                                  
```

```python
# subclass executing PCA before fitting the neural network. In this way larger neural netowrk can be computed at a much faster rate, but it might reduce accuracy.

class NN_PCA(NN):
    def __init__(self, min_explained_variance=0.90, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.min_explained_variance = min_explained_variance
        self.pca_ = None

    def fit(self, X, y):
        self.pca_ = PCA(n_components=self.min_explained_variance) # Apply PCA to the input data
        X_pca = self.pca_.fit_transform(X)
        return super().fit(X_pca, y)  # Call the fit method of the superclass

    def predict(self, X):
        X_pca = self.pca_.transform(X)   # Apply PCA to the input data
        return super().predict(X_pca)   # Call the predict method of the superclass

    def score(self, X, y):
        X_pca = self.pca_.transform(X)   # Apply PCA to the input data
        return super().score(X_pca, y)   # Call the score method of the superclass
```

### Building the Model Selection Function

Say that the best way to compare all these models is to use a function that efficiently comapres them

#### General overview

The following code is a series of functions that are used to construct the final function `sklearn_regressors_selector`. The script is specifically designed to efficiently manage the fitting and evaluation of ML regression models, in order to find the best model in the smallest amount of time possible.

**1. Differential handling of parallel and non-parallel models**: 
   - **Non-parallel models**: These models are fitted in parallel to fully utilize available computational resources, thereby speeding up the overall evaluation process. This is achieved by leveraging joblib's `Parallel` and `delayed` functions, which allow multiple non-parallel models to be fitted concurrently using all available processor cores.
   - **Parallel models**: Models that inherently support parallelization (i.e., those capable of using multiple cores for a single model fitting) are processed sequentially. This sequential processing ensures that the built-in parallel capabilities of these models are used effectively, avoiding the overhead and potential conflicts of external parallel execution.

**2. Preprocessing applied at each cross-validation**:
   - To avoid data leakage, the data preprocessing is applied independently to each train set within every cross-validation. The function shows information about the target and features data before and after preprocessing, as well as data on the models to be fitted.
     
**3. Sequential cross-validation with early stopping and timeout**:
   - The cross-validation process is kept sequential for each model. This allows us to implement an early stopping mechanism to save time and computational resources. If a model's performance in ongoing cross-validation is significantly lower than the best model so far (based on an adjustable margin), it can be prematurely skipped. You can also skip a model if its fitting exceeds a certain time period.

**4. Error handling and output options**:
   - **Error handling**: Errors during model instantiation or fitting are managed with informative messages about the nature of the failure.
   - **Output options**: The user has control over the complexity of the function's output. Specifically, you can choose to display general information, advanced details, or no information at all. Additionally, you can opt to show warnings, print the results table, and save the best model to disk.

**5. Results evaluation**:
   - After all models are processed, the results are sorted and displayed in a comprehensive table, showcasing each model's performance and status (e.g., success, timeout, skipped due to low performance, or failure in fitting / instantiation). Furthermore, there are shown for each model the averange of the R^2 obtained in the Cross Validation, their standard deviation, their median, as well as their average and median errors (both standardized and unstandardized).

#### What does every sub-function do?

Each function takes as input the previous function.
Here's a brief overview of each function within the script and their respective roles:

**1. fit_model_with_timeout**
- **Purpose**: Attempts to fit a ML model within a designated timeout for each cross-validation split.
- **Functionality**: It uses a thread pool to try fitting the model within the specified time limit for each split. If the model is successfully fitted within the timeout, it returns the model; otherwise, it returns `None` to signify a timeout.

**2. fit_model**
- **Purpose**: Fits a machine learning model using cross-validation, calculates the average R-squared across all splits, and manages parallelization flags.
- **Functionality**:
  - Initializes model based on whether it can be parallelized.
  - Processes each cross-validation split sequentially, fitting the model with a timeout and calculating the sum of squared residuals.
  - Evaluates if the model's performance is significantly worse than the best model so far and possibly skips the remaining splits.
  - Handles exceptions and errors during the fitting process, updating a results table with the status of each fitting attempt.
  - Determines if the current model has outperformed others in terms of R-squared and updates global best model variables accordingly.
  - Calculates standardized and unstandardized errors, as well as statistics on the R^2 results obtained for each model.
  - Optionally displays model parameters and any warnings encountered during fitting.

**3. parallel_fit**
- **Purpose**: Manages the execution strategy for fitting models by either parallelizing or processing them sequentially based on their capabilities.
- **Functionality**:
  - For models that are inherently parallelizable, fits them sequentially to maximize their built-in parallel processing capabilities.
  - For models that are not parallelizable, employs Joblib's `Parallel` and `delayed` functions to fit them in parallel.

**4. sklearn_regressors_selector**
- **Purpose**: Acts as the main driver function that orchestrates the selection and evaluation of regression models from scikit-learn, prints a results table, and identifies the best model based on cross-validation performance.
- **Functionality**:
  - Initializes cross-validation setup, applies data preprocessing to each train set, and optionally scales the data.
  - Displays info about the target, the features and the models.
  - Sorts models into parallelizable and non-parallelizable lists and executes model fitting using `parallel_fit` function for both lists of models.
  - Sorts and prints the results table of model evaluations.
  - Returns the best performing model and other relevant metrics to the user.

```python
# Import necessary libraries for data manipulation, machine learning, and parallel processing
# I am importing again libraries already imported because normally this function would be in
# its own file to be imported, but I am copy pasting it here instead, so I can show its code

import numpy as np
import pandas as pd
import random
import warnings
from sklearn.model_selection import KFold
from sklearn.utils import all_estimators
from sklearn.preprocessing import StandardScaler
from concurrent.futures import ThreadPoolExecutor, TimeoutError, as_completed
from joblib import Parallel, delayed, dump
import multiprocessing

# Define ANSI escape sequences for coloring terminal outputs
BLUE = '\033[38;5;19m' # Correct Results
DARK_YELLOW = '\033[38;5;3m' # Timeout
RED = '\033[0;31m' # Failed models
GREEN = '\033[0;32m' # Best Model
BOLD = '\033[1m' # Used for better contrast
ORANGE = '\033[38;5;208m' # Skipped models
RESET = '\033[0m' # Resetting normal formatting

# Initialize global variables that are set once, and then are constant and never change troughout the whole script
scale_data_flag = True
unstandardized_y = [] # used when scale_data_flag == True to obtain unstandardized residuals
standardized_y = [] # used when scale_data_flag == False to obtain standardized residuals
len_y_after_preprocessing = 0
original_means_sets_list = []
original_sd_sets_list = []
tss = None

# Initialize global variables that are meant to be continuosly modified by different functions
highest_R_squared = -np.inf
best_model = None
best_model_name = ""

# These two above are the only types of global variables included in this script. All other variables that could have 
# been global are instead just passed from function to function in order to preserve a high maintanability of the code

######################################################################################################################################
######################################################################################################################################

# This function attempts to fit a ML model using provided training data (X_train, y_train)
# within a specific timeout. This function will be used for fitting models whitin cross-validation
# Therefore, the timeout for each CV split is calculated by dividing the total allowed timeout by the number of CV splits

def fit_model_with_timeout(model, X_train, y_train, timeout, cv_splits): 
    with ThreadPoolExecutor() as executor:    # Attempt to fit a model within a specified timeout per CV split, using a thread pool executor for concurrency.
        future = executor.submit(model.fit, X_train, y_train) # Wait for model fitting to complete with a timeout adjusted for the number of CV splits.
        try:            
            future.result(timeout=timeout / cv_splits)
        except TimeoutError:
            return None  
    return model          # Return the updated model if the model fits successful within the timeout. If not, return None

# Important! In scikit-learn the fit method does not have a return value; it returns None. 
# The purpose of the fit method above is to modify the model object in-place. 
# After fitting, the model object itself contains the trained model parameters, which can be used for predictions

######################################################################################################################################
######################################################################################################################################

# The main role of function is to use 'fit_model_with_timeout' to fit a ML model using cross-validation on a provided dataset,
# and the calculating the average R squared across all the cross-validations.
# Additionally, it performs algorithms for optimization of time and processing resources, and for upgrading the final output displayed

# 1) It divides models between models that are parallelizable whitin themself and models that are not, flagging them.
# This will allow the next function 'parallel_fit' to fits in parallel between themself the non parallelizable models, 
# and then fits sequentially the ones already parallizable.
# In this way I achieves maximum time efficiency.

# 2) For every model, it cross validates them cv_split times. For all models, cross-validation is sequential, non parallel.
# In fact, the previous point allows for for maximum efficiency without the need of parallelizing the cross-validations.
# This makes possible to achieve point number 3

# 3) If a model performs poorly while cross validating, it skips it without the need to finish all the iterations.
# This is customizable: the user can set a margin treshold from the best model found so far.
# This allows to save time, avoiding wasting precious time and computation power for a model that would perform badly anyway.

# 4) It handles exceptions. They are both printed (with specific colors) and saved for the final results table.
# Models can fail instancing or fitting (red), can be skipped because of the margin treshold (orange), or can have a timeout (dark yellow)

# 5) It checks if the model is the best performing so far, and if so it saves it.

# 6) If enable, it displays model parameters and eventual warnings encountered while fitting. Warning are repeated once, to avoid having 1 warning for each CV.

# If the model performs well compared to previous models, it updates the global 'best_model' variables.
# Additionally, it can display model parameters and warnings if specified. The function appends results of each attempt (successful or not) to the 'results_table'.

def fit_model(sets_list, name, Model, kf, cv_splits, timeout, save_best_model_to_disk, 
              results_table, parallel_flag, display_warnings, display_models_parameters, 
              R_squared_margin_from_best_model):
    
    ssr_average = 0  # Sum of squared residuals accumulator
    model_warnings = set()   # Set to accumulate unique warnings per model
    model_errors = 0   # Counter for tracking model fit failures
    global len_y_after_preprocessing
    abs_tot_residuals = np.empty(len_y_after_preprocessing)
    tot_pred = []
    residuals_index = 0
    ssr_list = np.empty(cv_splits) # keeps SSR data to compute variance, minimum and maximum

    # Declare use of global variables. These variables must be global so they can be constantly updated by every parallel process without compromising other processes 
    global highest_R_squared
    global best_model
    global best_model_name
    global scale_data_flag
    global unstandardized_y # used when scale_data_flag == True to obtain unstandardized residuals
    global standardized_y # used when scale_data_flag == False to obtain standardized residuals
    global original_means_sets_list
    global original_sd_sets_list
    
    try:
        if parallel_flag: # If the model can be parallelized, do so using the maximum amount of threads (use all processor cores + eventual hyperthreads within a processor)
            model = Model(n_jobs=-1) # Reminder: the process of fitting multiple parallel models will not be parallelized but completed sequentially
        else:             # If it cannot be parallelized, fit the model sequentially.
            model = Model()    # Reminder: the process of fitting multiple non-parallel models will be parallelized.

    # Handle exceptions during model instantiation (commonly due to meta estimators as they require additional parameters in order to be fitted)
    # For every model either its final result or its ecountered error is the appended to the results table, which will be displayed after all the models are cross-validated
    # The models that fail instantiation should have already been removed beforehand, as the main function sklearn_regressors_selector removes them when dividing parallel and not parallel models
    # Anyway, this try and except has been kept to maintain stability in case of unexpected errors
    except Exception as e:
        print(f"{RED}{name:<35} Failed instantiating, model is likely a meta estimator: {RESET}{RED} {str(e)}{RESET}")
        results_table.append([name, 'FAIL - instantiating','','','','','','','',''])
        return

    with warnings.catch_warnings(record=True) as w:
        warnings.simplefilter("always") # Configure to capture warnings once per unique warning
        
        cv_counter = 0   # Counter for tracking the number of completed CV splits

        for X_train, X_test, y_train, y_test, in sets_list:   # CV loop
            fitted_model = None
            cv_counter += 1
            
            try:                      # Fit the model with a timeout
                fitted_model = fit_model_with_timeout(model, X_train, y_train, timeout, cv_splits)
                if fitted_model is None:
                    print(f"{DARK_YELLOW}{name:<35} TIMEOUT{RESET}")
                    model_errors += 1
                    results_table.append([name, 'TIMEOUT','','','','','','','',''])     # Handle exceptions
                    break
                
                # Predict using the fitted model and calculate the sum of squared residuals for this split.
                y_pred = fitted_model.predict(X_test)
                residuals_current = y_test - y_pred
                ssr_current = np.sum(residuals_current ** 2)
                ssr_average += ssr_current
                ssr_list[cv_counter - 1] = ssr_current
                
                # Saving residuals data and predictions data
                number_of_residuals = len(residuals_current)
                abs_tot_residuals[residuals_index:residuals_index+number_of_residuals] = np.abs(residuals_current) # keeps the total number of residuals, used to know the average and median residual
                tot_pred.append(y_pred)
                residuals_index += number_of_residuals

                # Calculate the adjusted R-squared and check if it is significantly worse than the best model (R_squared_margin_from_best_model, parameter set by the user)
                # If that is the case, avoid wasting time doing other cross-validations, and skip to the next model
                scaled_R_squared = 1 - (ssr_average * (cv_splits / cv_counter) / tss)
                if scaled_R_squared < highest_R_squared - R_squared_margin_from_best_model:
                    print(f"{ORANGE}{name:<35} Skipped, CV R^2 too low. CV R^2: {scaled_R_squared * 100:.2f}% vs best CV R^2: {highest_R_squared * 100:.2f}%{RESET}")
                    results_table.append([name, 'SKIPPED - Low CV R^2','','','','','','','',''])     
                    model_errors += 1
                    break
            except Exception as e:    # Handle exceptions
                print(f"{RED}{name:<35} Model fit error:{RESET}{RED} {str(e)}{RESET}")
                model_errors += 1
                results_table.append([name, 'FAIL - Fitting','','','','','','','',''])
                break

        # If a model completed all CV splits without errors and wasn't skipped, print its score and append it to the results table
        # also, append to the table other helpful statistics, which are:
        # for R2: standard deviation, mean and median
        # for residuals: mean, median, standardized mean, standardized median
        if fitted_model and not model_errors:
            R_squared = 1 - (ssr_average / tss)
            ssr_list = 1 - (ssr_list / tss * cv_splits) # this list is now actually a R_squared list. I did not create a new variable because it is a slower (but clearer) method
            sd_R_squared = np.std(ssr_list)
            median_R_squared = np.median(ssr_list)
            min_R_squared = np.min(ssr_list)
            max_R_squared = np.max(ssr_list)
            
            # getting standardized and unstandardized residual data
            if scale_data_flag: # if y_pred and abs_tot_residuals are already standardized, get unstandardized residuals
                unstandardized_y_pred = np.empty(len_y_after_preprocessing)
                unstandardized_y_index = 0
                
                for i in range(cv_splits):
                    number_of_unstandardized_y = len(tot_pred[i])
                    unstandardized_y_pred[unstandardized_y_index:unstandardized_y_index+number_of_unstandardized_y] = (
                        tot_pred[i] * original_sd_sets_list[i] + original_means_sets_list[i])
                    unstandardized_y_index += number_of_unstandardized_y
                
                abs_unstandardized_residuals = np.abs(unstandardized_y - unstandardized_y_pred)
                mean_residuals = np.mean(abs_unstandardized_residuals)
                median_residuals = np.median(abs_unstandardized_residuals)
                st_mean_residuals = np.mean(abs_tot_residuals)
                st_median_residuals = np.median(abs_tot_residuals)
                    
            else: # if y_pred and abs_tot_residuals are unstandardized, get standardized residuals
                standardized_y_pred = np.empty(len_y_after_preprocessing)
                standardized_y_index = 0

                for i in range(cv_splits):
                    number_of_standardized_y = len(tot_pred[i])
                    standardized_y_pred[standardized_y_index:standardized_y_index+number_of_standardized_y] = (
                    (tot_pred[i] - original_means_sets_list[i]) / original_sd_sets_list[i])
                    standardized_y_index += number_of_standardized_y
                    
                abs_standardized_residuals = np.abs(standardized_y - standardized_y_pred)
                mean_residuals = np.mean(abs_tot_residuals)
                median_residuals = np.median(abs_tot_residuals)
                st_mean_residuals = np.mean(abs_standardized_residuals)
                st_median_residuals = np.median(abs_standardized_residuals)
    
            print(f"{BLUE}{name:<35} {R_squared * 100:.2f}%{RESET}")
            results_table.append([name,R_squared,sd_R_squared,median_R_squared, min_R_squared, max_R_squared,
                                  mean_residuals, median_residuals, st_mean_residuals, st_median_residuals])
            
        # Is the current model the best so far?
            if R_squared > highest_R_squared:
                highest_R_squared = R_squared
                best_model = fitted_model
                best_model_name = name

                # save best model so far to disk
                if save_best_model_to_disk:
                    dump(fitted_model, 'best_model.pkl')

        # Display model parameters and warnings if enabled
        if display_models_parameters:
            print(f'{fitted_model.get_params()} \n')
    
        if display_warnings:
            for warning in w:
                model_warnings.add(str(warning.message))
            if model_warnings:
                for warning in model_warnings:
                    if display_models_parameters:
                        print(f"{RESET}{BOLD}Warning - {name}: {warning}{RESET}\n")
                    else:
                        print(f"{RESET}Warning - {name}: {warning}\n")

######################################################################################################################################
######################################################################################################################################

# This function fits sequentially the models flagged with the parallel flag, and fits in parallel the models not flagged.
# To fit the models it uses the function above 'fit_model'.
# To achieve parallelization it uses joblib's 'Parallel' function in combination with 'delayed'

def parallel_fit(sets_list, model_list, kf, cv_splits, timeout, save_best_model_to_disk, 
                 parallel_flag, results_table, display_warnings, display_models_parameters, 
                 R_squared_margin_from_best_model):
    # Already parallelized models
    if parallel_flag:
        for name, Model in model_list:
            fit_model(sets_list, name, Model, kf, cv_splits, 
                    timeout, save_best_model_to_disk, results_table, parallel_flag, 
                    display_warnings, display_models_parameters, R_squared_margin_from_best_model)

    # Non - parallel models. Using Joblib parallel function to achieve multithreading
    # n_jobs specifies the number of parallel jobs to run
    # n_jobs = -1 uses all available cores on the machine (use all processor cores + eventual hyperthreads within a processor)
    # Using function 'delayed' in combination with 'Parallel'. delayed "wraps" a function, storing the function itself 
    # and the arguments you want to pass to it. This wrapped function will only execute when explicitly told to, not when it's initially set up.
    # By wrapping fit_model with delayed, each call to fit_model is turned into a task that can be executed as a separate job. 
    # This allows Parallel to manage the distribution of these tasks across different threads or processes based on the specified backend.
    # The backend parameter specifies the parallel execution strategy and how the tasks are managed and executed in multiple threads or processes

    else: 
        Parallel(n_jobs=multiprocessing.cpu_count(), backend='threading')(
            delayed(fit_model)(
                sets_list, name, Model, kf, cv_splits, timeout, save_best_model_to_disk, results_table, 
                parallel_flag, display_warnings, display_models_parameters, R_squared_margin_from_best_model
            )for name, Model in model_list  # Iterating over the function models_list which contains tuples of (name, Model)
        )

# When running joblib function 'parrallel' in Jupyterlab, the only backend working is 'threading'. Multiprocessing crashes immediately, and loky does not print outputs and blocks after a while.
# Therefore, if you run this code outside jupyterlab, try different backends, you might benefit more from them than from 'multiprocessing'
# This because 'threading' does not use separate memory spaces and does not release the GIL, requiring the functions passed trough it to release it themself.
# The GIL (Global Interpreter Lock) is a mechanism used in standard Python implementation (CPython) to ensure that only one thread executes Python bytecode at a time.
# This limits processes to be single-threaded. This slows CPU-bound tasks significantly, but it is set as default since CPython's memory management is not thread-safe by itself
# If you release the GIL, make sure you use libraries that ensure thread safeness.
# In our case, the interactive nature of jupyterlab cells seems to be a problem when choosing the backend. Does using backend = 'threading' limit parallelization?
# I monitored the CPU usage when running the code (laptop charging and in 'high performance' power setting) and it seems that CPU utilizations levels are often sustained above 60-70%, and reach
# 100% utilization multiple times (in theory I would hope that usage is alwasy close to 100%, but that is not always possible as the function delayed needs to make sure that everything
# is safe and coordinated before letting threads run new computational jobs). This result means that parallelization is achieved in the great majority of the running time of the script.

######################################################################################################################################
######################################################################################################################################

# Main function to select the best performing sklearn regressor model based on cross-validation R-squared
# It also prints the results table, fits again the best model and output is out-of-CV R squared,
# and finally it returns useful variables to the user
def sklearn_regressors_selector(X, y, preprocessing_function = None, models = all_estimators(type_filter='regressor'),
                                cv_splits = 5, random_seed=random.randint(1,10000),
                                timeout = 300, scale_data = True, save_best_model_to_disk = False,
                                display_warnings = True, display_models_parameters = False, display_general_info = True,
                                display_advanced_info = False, print_results_table = True, R_squared_margin_from_best_model = np.inf):

    # importing global variables in order to modify them
    global scale_data_flag
    global unstandardized_y # used when scale_data_flag == True to obtain unstandardized residuals
    global standardized_y # used when scale_data_flag == False to obtain standardized residuals
    global original_means_sets_list
    global original_sd_sets_list
    global len_y_after_preprocessing
    global tss

    # convert to panda. This is needed to apply preprocessing.
    # Later, our data will be converted to numpy to speed up calculations
    if not isinstance(X, pd.DataFrame):
        try:
            X = pd.DataFrame(X)
        except Exception as conversion_X_to_panda_DF_error:
            return str(conversion_X_to_panda_DF_error)

    if not isinstance(y,pd.Series):
        try:
            y = pd.Series(y)
        except Exception as conversion_y_to_panda_Series_error:
            return str(conversion_y_to_panda_Series_error)

    # printing disclaimer in case you want to save the best model
    if save_best_model_to_disk:
        print(f"\n{RED}{BOLD}DISCLAIMER - SAVING BEST MODEL")
        print("\nIt is recommended to run your model individually when using this function to save it.")
        print("This is because the best model is dynamically saved each time a model outperforms its predecessors.")
        print("Using the 'save_best_model_to_disk' parameter while fitting multiple models can cause errors,") 
        print(f"as one model might overwrite the 'best_model.pkl' file while another is still in the process of saving.{RESET}")
        print(f"\n{RED}Additionally, utilize 'save_best_model_to_disk' solely for gaining deeper insights into the model's operations.")
        print("Avoid employing the saved model for fitting purposes, as you should train with the entire training dataset before testing.")
        print(f"The saved model was trained with only {round(1 - 1/cv_splits, 3)} of the available training data due to cross-validation.{RESET}")

    # printing features info before preprocessing
    if display_general_info:
        if preprocessing_function == None:
            print(f'\n{BOLD}FEATURES{RESET}')
        else:
            print(f'\n{BOLD}FEATURES BEFORE PREPROCESSING{RESET}')

        num_of_features = X.shape[1]
        print(f'Number of features: {num_of_features}')
        dummy_columns = X.columns[(X.nunique() == 2) & (X.isin([0, 1]).all())]
        dummy_columns_len = len(dummy_columns)
        print(f'of which dummies: {dummy_columns_len}')
        print(f'of which non-dummies: {num_of_features - dummy_columns_len}')
        if display_advanced_info:
            
            # For better aesthetics, I define a function that prints the list of dummies in lines that are no longer than 135 characters
            def print_list_within_len_limit(data_list, line_length=135):
                current_line = "" # Initialize an empty string to build the current line
                # Check if adding the next item would exceed the specified line length
                for item in data_list:
                    if len(current_line) + len(item) + 2 > line_length:  # adding +2 for the comma and space between dummies: ", "
                        print(current_line[:-2])  # print current line removing the string ", " at the end
                        current_line = ""
                    current_line += item + ", "
                if current_line:  # print the remaining items in the last line, and remove the comma after the last element
                    print(current_line[:-2])
            
            print(f'\n{BOLD}DATA TYPES OF FEATURES{RESET}')
            print(f'{X.dtypes.value_counts()}')
            print(f'\n{BOLD}LIST OF DUMMIES{RESET}')          
            print_list_within_len_limit(dummy_columns.tolist())
            
    # printing target info before preprocessing
        if preprocessing_function is None:
            print(f'\n{BOLD}TARGET{RESET}')
        else:
            print(f'\n{BOLD}TARGET BEFORE PREPROCESSING{RESET}')
            
        print(f'Mean: {y.mean():.2f}')
        print(f'Median: {y.median():.2f}')
        print(f'SD: {y.std():.2f}')
        print(f'Min: {y.min():.2f}')
        print(f'Max: {y.max():.2f}')
        print(f'Observations: {len(y)}')
        print(f"\n{RESET}---------------------------------------------------------------------------------------------------------------------------------------") 

    # The scale_data_flag is used to signal that data are standardized, and therefore
    # non standardized errors will have to be calculated in the fit_model function
    # Viceversa, if the flag is false, standardized errors will be calculated.
    # The scalers_list is used to collact all the scalers used to calculate the standardized_y
    if not scale_data:
        scale_data_flag = False
        scalers_list = []    

    # split the data for cross-validation
    kf = KFold(n_splits=cv_splits, shuffle=True, random_state=random_seed) # Initialize KFold with random shuffling
    sets_list = [] # create list with all train and test sets after you have applied the preprocessing.
    tss_sets_list = []
    did_we_display_preprocessed_info = False    # used to display preprocessed info just once
    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]

        # apply preprocessing
        if preprocessing_function != None:
            X_train, y_train = preprocessing_function(X_train,y_train)
            X_test, y_test = preprocessing_function(X_test,y_test)
            
            # Ensure test data has the same columns as training data after preprocessing
            train_columns = set(X_train.columns)
            test_columns = set(X_test.columns)
            common_columns = list(train_columns & test_columns) # Columns present in both
            missing_columns = list(train_columns - test_columns) # Columns missing in test set
            X_test = X_test[common_columns] # Drop extra columns from test set
            for col in missing_columns:  # Add missing columns with default value of 0
                X_test[col] = 0
            X_test = X_test[X_train.columns] # Ensure the order of columns matches training data

            # print info about the features after they have been preprocessed
            # To avoid printing too much, only the information about the first X_train set is printed
            # This because feature info are unlikely to change much between different train sets
            if not did_we_display_preprocessed_info and display_general_info:
                did_we_display_preprocessed_info = True
                print(f'\n{BOLD}FEATURES AFTER PREPROCESSING{RESET}')
                print('(Statistics obtained using only the first training set from cross-validation)')
                num_of_features_after_preproc = X_train.shape[1]
                print(f'Number of features: {num_of_features_after_preproc}')
                dummy_columns_after_preproc = X_train.columns[(X_train.nunique() == 2) & (X_train.isin([0, 1]).all())]
                len_dummy_columns_after_preproc = len(dummy_columns_after_preproc)
                print(f'of which dummies: {len_dummy_columns_after_preproc}')
                print(f'of which non-dummies: {num_of_features_after_preproc - len_dummy_columns_after_preproc}')
                if display_advanced_info:
                    print(f'\n{BOLD}DATA TYPES OF FEATURES{RESET}')
                    print(f'{X_train.dtypes.value_counts()}')
                    print(f'\n{BOLD}LIST OF DUMMIES{RESET}')
                    print_list_within_len_limit(dummy_columns_after_preproc.tolist())

                # Similarly, print info about the features after they have been preprocessed
                # The total number of observations after preprocessing will be printed outside the loop, to reflect the real number and not just an estimate
                print(f'\n{BOLD}TARGET AFTER PREPROCESSING{RESET}')
                print('(Statistics obtained using only the first training set from cross-validation, except observations)')
                print(f'Mean: {np.mean(y_train):.2f}')
                print(f'Median: {np.median(y_train):.2f}')
                print(f'SD: {np.std(y_train):.2f}')
                print(f'Min: {np.min(y_train):.2f}')
                print(f'Max: {np.max(y_train):.2f}')

        # I scale the data and convert it into the right numpy array
        # for X data I change first to numpy array, then I apply the scaler
        # For y, it needs to be first transformed to a 2D numpy, then applied the scaler, then converted to 1D numpy  
        # Furthermore, all the standardized y are collected before the standardization is applied.
        # This global variable will be used to calculate unstandardized errors
        if scale_data:
            unstandardized_y.extend(y_test.values.reshape(-1))
            scaler_X = StandardScaler()  #initialize scalers
            scaler_y = StandardScaler()
            
            X_train = np.array(X_train) #convert to numpy
            X_test = np.array(X_test)
            y_train = y_train.values.reshape(-1, 1) # convert Panda series to 2D arrays to apply scalers
            y_test = y_test.values.reshape(-1, 1)
            
            scaler_X.fit(X_train) # fit scalers only on training data
            scaler_y.fit(y_train)
            
            X_train = scaler_X.transform(X_train)    # apply scalers
            X_test = scaler_X.transform(X_test)
            y_train = scaler_y.transform(y_train)
            y_test = scaler_y.transform(y_test)
            
            y_train = y_train.reshape(-1)   # converting from 2D numpy to 1D numpy
            y_test = y_test.reshape(-1)
     
        else:
            # For the non-scaling case, I do similar transformations as above
            # Also, I generate and collect the standardized_y
            
            # populating the standardized_y list
            scaler_y = StandardScaler()
            y_train = y_train.values.reshape(-1, 1)
            y_test = y_test.values.reshape(-1, 1)
            scaler_y.fit(y_train)
            standardized_y.extend(scaler_y.transform(y_test).reshape(-1))
            
            # converting to numpy
            X_train = np.array(X_train)
            X_test = np.array(X_test)
            y_train = y_train.reshape(-1)
            y_test = y_test.reshape(-1)

        # Finally, all numpy arrays will be made C-contiguous and aligned, to maximize speed of fitting.
        # C-contiguous means that the elements are stored in a contiguous block of memory,
        # and also that they are stored row-wise, which is useful since fitting models follows a row-wise pattern.
        # This configuration is set by default when creating a numpy array, but slicing can compromise this propriety
        # Aligned refers to the memory addresses where the data is stored. 
        # Proper alignment can improve performance because it matches the hardware's expected data boundaries, 
        # reducing the number of memory access operations. Also alignment is set by default, but is compromized by slicing.
        X_train = np.require(X_train, requirements=['C', 'A']) # making the arrays contiguous and aligned
        X_test = np.require(X_test, requirements=['C', 'A'])
        y_train = np.require(y_train, requirements=['C', 'A'])
        y_test = np.require(y_test, requirements=['C', 'A'])

        # I finally save several lists and variable that will be used
        # inside fit_model when calculating the unstandardized / standardized errors
        # These variables are the mean of the sample, the standard deviation and the amount of elements of y
        original_means_sets_list.append(scaler_y.mean_)
        original_sd_sets_list.append(np.sqrt(scaler_y.var_))
        tss_sets_list.append(np.sum((y_test - np.mean(y_test)) ** 2))
        len_y_after_preprocessing += len(y_test)

        # I save a list for each loop, each containing a cross-validation split (X_train,X_test,y_train,y_test)
        # They are used in fit_model to train and test within CV
        sets_list.append([X_train,X_test,y_train,y_test])

    # Total sum of squares. Used to calculate the R2 in fit_model
    tss = np.sum(np.array(tss_sets_list))
    
    # print number of y observations after preprocessing
    if preprocessing_function != None and display_general_info:
        print(f'Observations: {len_y_after_preprocessing}')
        print(f"\n{RESET}---------------------------------------------------------------------------------------------------------------------------------------") 

    # Initializing the results table
    results_table = []

    # Classify models into parallelizable and non-parallelizable based on 'n_jobs' parameter presence
    non_parallelizable = []
    parallelizable = []
    instantiating_error = []
    
    for name, Model in models:
        try:
            params = Model().get_params()
            if 'n_jobs' in params:
                parallelizable.append((name, Model))
            else:
                non_parallelizable.append((name, Model))
        except:
            instantiating_error.append((name, Model))
            results_table.append([name, 'FAIL - instantiating','','','','','','','',''])

    if display_general_info:
        # printing models informations
        print(f'\n{BOLD}MODELS{RESET}')
        print(f'Number of models: {len(models)}')
        print(f'Non parallel: {len(non_parallelizable)}')
        print(f'Parallel: {len(parallelizable)}')
        inst_err_len = len(instantiating_error)
        if inst_err_len > 0:
            print(f'Instantiating error \n(requiring more parameters, likely meta estimators): {inst_err_len}')
        if display_advanced_info:
            print(f'\n{BOLD}LIST OF NON PARALLEL MODELS{RESET}')
            for name_of_model, parameters_of_model in non_parallelizable:
                print(name_of_model)
            print(f'\n{BOLD}LIST OF PARALLEL MODELS{RESET}')
            for name_of_model, parameters_of_model in parallelizable:
                print(name_of_model)
            if inst_err_len > 0:
                print(f'\n{BOLD}LIST OF INSTANTIATING-ERROR MODELS{RESET}')
                for name_of_model, parameters_of_model in instantiating_error:
                    print(name_of_model)

        # printing function informations
        print(f'\n{BOLD}FUNCTION{RESET}')
        print(f'CV folds: {cv_splits}')
        print(f'Seed: {random_seed}')
        if timeout > 10 * cv_splits:
            print(f'Timeout: {timeout} sec, {(timeout / cv_splits):.0f} for each CV')
        else:
            print(f'Timeout: {timeout} sec, {(timeout / cv_splits):.2f} for each CV')
        print(f'Scale data: {scale_data}')
        print(f'Margin from best model: {R_squared_margin_from_best_model * 100}%\n')
        
    print(f"{RESET}---------------------------------------------------------------------------------------------------------------------------------------")    
    print(f"\n{BOLD}MODEL                               CV R^2{RESET}") # Printing output header

    # Executing first the non_parallelizable models, and the the parallelizable models
    models_lists = [non_parallelizable, parallelizable]
    for model_list in models_lists:
        parallel_flag = False
        if model_list == parallelizable:
            parallel_flag = True

        # Using the parallel_fit function (remember that it fits in parallel non-parallel models and sequentially the already parallelized ones)
        # For this reason, I pass the parallel_flag to it
        parallel_fit(sets_list, model_list, kf, cv_splits, timeout, save_best_model_to_disk,
                     parallel_flag, results_table, display_warnings, display_models_parameters,
                     R_squared_margin_from_best_model)

    # Sort results table by performance and error type for better visualization
    results_table.sort(key=lambda item: (0, -item[1]) if isinstance(item[1], float) else 
                                       (1,) if item[1] == "TIMEOUT" else 
                                       (2,) if item[1] == "SKIPPED - Low CV R^2" else
                                       (3,) if item[1] == "FAIL - fitting" else 
                                       (4,) if item[1] == "FAIL - instantiating" else 
                                       (5,))

    # Print the sorted results table if enabled
    if print_results_table:
        print(f"\n{RESET}---------------------------------------------------------------------------------------------------------------------------------------\n")
        print(f'{BOLD}{"RESULTS TABLE":<35} {"CV R^2":<10}  {"SD":<10} {"MED":<10} {"MIN":<10} {"MAX":<10}{"AV |ER|":<9}  {"MED |ER|":<10} {"ST AV |ER|":<10} {"ST MED |ER|":<10}{RESET}')
        for model_name, value, sd, median, minimum, maximum, mean_res, median_res, st_mean_res, st_median_res in results_table:
            color_code = BLUE if isinstance(value, float) else DARK_YELLOW if 'TIMEOUT' in value else (ORANGE if 'SKIPPED - Low CV R^2' in value else RED)  
            
            value_display = f"{value * 100:.2f}%" if isinstance(value, float) else value
            sd_display = f"{sd * 100:.2f}%" if isinstance(sd, float) else ''
            median_display = f"{median * 100:.2f}%" if isinstance(median, float) else ''
            min_display = f"{minimum * 100:.2f}%" if isinstance(minimum, float) else ''
            max_display = f"{maximum * 100:.2f}%" if isinstance(maximum, float) else ''
            mean_res_display = f"{mean_res:.2f}" if isinstance(mean_res, float) else ''
            median_res_display = f"{median_res:.2f}" if isinstance(median_res, float) else ''
            st_mean_res_display = f"{st_mean_res:.2f}" if isinstance(st_mean_res, float) else ''
            st_median_res_display = f"{st_median_res:.2f}" if isinstance(st_median_res, float) else ''

            print(
                    f"{color_code}{model_name:<35} {value_display:<10} "
                    f"{sd_display:<10} {median_display:<10} "
                    f"{min_display:<10} {max_display:<10} "
                    f"{mean_res_display:<8}    {median_res_display:<10} "
                    f"  {st_mean_res_display:<10} {st_median_res_display:<10}{RESET}"
                 )
            
    # Note how I import now global variables, since their value has changed after I executed the function 'parallel_fit'
    global best_model
    global best_model_name
    global highest_R_squared

    # Output the final best model performance and related statistics
    print(f"\n{RESET}---------------------------------------------------------------------------------------------------------------------------------------\n")
    print(f"{BOLD}{'BEST MODEL:':<{35}} {GREEN}{BOLD}{best_model_name}{RESET}")
    print(f"{'CV R^2:':<{35}} {GREEN}{highest_R_squared * 100:.4f}%{RESET}")

    # Reset global variables in preparation for potential future function calls, as this function is not imported from a file 
    # but just pasted in a jupyter notebook cell, and therefore global variables do not reset every time this function is called
    best_performing_model_name = best_model_name  # these next passages are done since I am in a jupyterlab cell, and I need to reset parameters
    best_performing_model = best_model # normally, I would avoid it and just return best_model and best_model_name
    best_model = None 
    best_model_name = ""

    best_model_R_squared = highest_R_squared  # similarly, all the following commands would normally not be necessary
    highest_R_squared = -np.inf 
    scale_data_flag = True
    unstandardized_y = []
    standardized_y = []
    len_y_after_preprocessing = 0
    original_means_sets_list = []
    original_sd_sets_list = []
    tss = None

    # return useful variables to the user
    return best_model_R_squared, best_performing_model_name, best_performing_model, results_table
```

### Hyperparameter Tuning with Nested CV

- **What is hyperparameter tuning**
  
Hyperparameter tuning is the process of finding the best configuration settings for a machine learning model. These settings, known as hyperparameters, are set before the training process begins and can significantly impact the model's performance. Examples include the learning rate in gradient descent or the depth of a decision tree. The goal is to identify the hyperparameters that provide the best performance on a validation dataset.

- **What is nested CV**

Nested cross-validation is a robust technique for hyperparameter tuning and performance evaluation. It involves two layers of cross-validation: an outer loop and an inner loop. The outer loop splits the data into training and testing sets multiple times to estimate the model's performance. Within each of these training sets, the inner loop performs another level of cross-validation to tune the hyperparameters by splitting the training data further into smaller training and validation sets.

- **Why should you use nested CV with hyperparameter tuning**

Using nested cross-validation ensures an unbiased estimate of model performance because the test set in the outer loop remains completely independent of the hyperparameter tuning process in the inner loop. This method reduces the risk of overfitting to the validation data and provides a more accurate assessment of how the model will perform on new, unseen data.

- Firstly, I make a custom function for hyperparameter tuning inside a nested CV.
- Secondly, I create an hyperparameter space for each model.
- Then, I do hyperparameter tuning on the best 6 models. 
- Finally, I select the best one with its respective hyperparameters, and I tune again using hyperparameters with values around the ones found.

I created our nested cross-validation function for hyperparameter tuning.

```python
from sklearn.model_selection import KFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import numpy as np
import time

def HT_with_nested_CV(models, param_distributions, X_train, y_train, preprocessing_function, 
                            outer_cv_splits = 3, inner_cv_splits = 3, random_search_iter = 50, random_search_random_state = random.randint(1,10000),
                           random_search_scoring = 'neg_mean_squared_error'):
    # Define ANSI escape sequences for coloring terminal outputs
    BLUE = '\033[38;5;19m'  # Correct Results
    RED = '\033[0;31m'  # Failed models
    GREEN = '\033[0;32m'  # Best Model
    BOLD = '\033[1m'  # Used for better contrast
    RESET = '\033[0m'  # Resetting normal formatting

    # Define outer and inner cross-validation
    outer_cv = KFold(n_splits=outer_cv_splits)
    inner_cv = KFold(n_splits=inner_cv_splits)

    outer_scores = {name: [] for name in models.keys()}
    best_hyperparameters = {name: [] for name in models.keys()}

    # Outer loop
    for i, (train_idx, test_idx) in enumerate(outer_cv.split(X_train, y_train)):
        print(f"\n{RESET}---------------------------------------------------------------------------------------------------------------------------------------")
        print(f'\n {RED}{BOLD}Outer fold {i+1}/{outer_cv.get_n_splits()}{RESET}')
        X_train_nested, X_test_nested = X_train.iloc[train_idx], X_train.iloc[test_idx]
        y_train_nested, y_test_nested = y_train.iloc[train_idx], y_train.iloc[test_idx]

        # Apply preprocessing function to the outer training set
        X_train_nested, y_train_nested = preprocessing_function(X_train_nested, y_train_nested)
        X_test_nested, y_test_nested = preprocessing_function(X_test_nested, y_test_nested)

        # Ensure test data has the same columns as training data after preprocessing
        train_columns = set(X_train_nested.columns)
        test_columns = set(X_test_nested.columns)
        common_columns = list(train_columns & test_columns)  # Columns present in both
        missing_columns = list(train_columns - test_columns)  # Columns missing in test set
        X_test_nested = X_test_nested[common_columns]  # Drop extra columns from test set
        for col in missing_columns:  # Add missing columns with default value of 0
            X_test_nested[col] = 0
        X_test_nested = X_test_nested[X_train_nested.columns]  # Ensure the order of columns matches training data

        scaler_X = StandardScaler()  # Initialize scalers
        scaler_y = StandardScaler()

        X_train_nested = np.array(X_train_nested)  # Convert to numpy
        X_test_nested = np.array(X_test_nested)
        y_train_nested = y_train_nested.values.reshape(-1, 1)  # Convert Panda series to 2D arrays to apply scalers
        y_test_nested = y_test_nested.values.reshape(-1, 1)

        scaler_X.fit(X_train_nested)  # Fit scalers only on training data
        scaler_y.fit(y_train_nested)

        X_train_nested = scaler_X.transform(X_train_nested)  # Apply scalers
        X_test_nested = scaler_X.transform(X_test_nested)
        y_train_nested = scaler_y.transform(y_train_nested)
        y_test_nested = scaler_y.transform(y_test_nested)

        y_train_nested = y_train_nested.reshape(-1)  # Converting from 2D numpy to 1D numpy
        y_test_nested = y_test_nested.reshape(-1)

        for model_name, model in models.items():
            start_time_inner = time.perf_counter()
            # Inner loop: Hyperparameter tuning
            print(f'\n  {BLUE}Tuning {GREEN}{model_name}{BLUE}...{RESET}')
            random_search = RandomizedSearchCV(estimator=model, param_distributions=param_distributions[model_name],
                                               n_iter=random_search_iter, scoring=random_search_scoring, return_train_score=False,
                                               cv=inner_cv, verbose=2, random_state=random_search_random_state, n_jobs=-1)

            random_search.fit(X_train_nested, y_train_nested)

            # Best model from inner loop
            best_model = random_search.best_estimator_

            best_hyperparameters[model_name].append(random_search.best_params_)
            print(f'    {BLUE}Best parameters: {random_search.best_params_}{RESET}')

            # Evaluate best model on outer test set
            y_pred_nested = best_model.predict(X_test_nested)
            outer_score = r2_score(y_test_nested, y_pred_nested)
            print(f'    {BLUE}R squared: {BOLD}{outer_score:.4f}{RESET}')
            outer_scores[model_name].append(outer_score)
            end_time_inner = time.perf_counter()
            print(f"Executed in {end_time_inner - start_time_inner:.2f} sec")

    # Final performance
    print(f"\n{RESET}---------------------------------------------------------------------------------------------------------------------------------------")
    print(f'\n{RED}{BOLD}FINAL PERFORMANCE{RESET}')

    average_scores = {name: np.mean(scores) for name, scores in outer_scores.items()}

    for name, score in average_scores.items():
        print(f'\n{GREEN}{BOLD}{name}{RESET}')
        print(f'\nAverage R squared: {BLUE}{BOLD}{score:.4f}{RESET}')

        # Create a set to keep track of unique hyperparameters
        best_hyperparameters_without_repetitions = set()
        # Iterate through the list of best hyperparameters for the current model
        for index, hyperparameters in enumerate(best_hyperparameters[name]):
            # Convert hyperparameters dictionary to a tuple for set operations
            hyperparameters_tuple = tuple(hyperparameters.items())
            # Check if the hyperparameters tuple is not in the set to avoid repetitions
            if hyperparameters_tuple not in best_hyperparameters_without_repetitions:
                # Add the new unique hyperparameters tuple to the set
                best_hyperparameters_without_repetitions.add(hyperparameters_tuple)
                # Print the indexes of the outer folds where the current hyperparameters were best
                print(f"\n Best hyperparameters for outer folds {BLUE}{[i for i, hy in enumerate(best_hyperparameters[name]) if hy == hyperparameters]}{RESET}:")
                # Print each hyperparameter key and its value in blue and bold
                for key, value in hyperparameters.items():
                    print(f"{key}: {BLUE}{BOLD}{value}{RESET}")

    return average_scores, best_hyperparameters
```

## What is the most unbiased and efficient way to compute the best predictive algorithm?

To identify the most unbiased and efficient way to compute the best predictive algorithm, I employed a two-step approach. 

Firstly, I compared a variety of models, including all 54 regressor models in scikit-learn, 8 custom PyTorch neural networks, XGBoost, and LightGBM, using 5-fold cross-validation. This was performed on a dataset comprising 13.188 observations and around 90 features (the number of features changes depending on the training set). The process was executed on a Thinkpad E14 gen 5 with intel i7 13-700h without a GPU. The process took under 180 seconds, demonstrating the efficiency of such a function.

Then, I tuned their hyperparameter within a nested Cross Validation. This method provides an unbiased estimate of model performance by ensuring the test set in the outer loop remains independent of the hyperparameter tuning process, minimizing overfitting risks.

**1. Model Comparison and Initial Evaluation**: 
   
The `sklearn_regressors_selector` function manages the initial fitting and evaluation of regression models. It uses joblib's Parallel and delayed functions to perform parallel processing for models that lack built-in parallelism. For models with inherent parallel capabilities, the function processes them sequentially to optimize their built-in efficient management of computational resources. 

Each cross-validation fold is independently preprocessed to prevent data leakage. Despite common practice, cross validations of the same model are fitted sequentially. This approach allows the function to implement an early stopping mechanism. If a model's performance during cross validation significantly lags behind the best model so far, the function can terminate it early, avoiding the need to fit the remaining splits and conserving computational resources. Furthermore, the function also implements a timeout.
   
**2. Hyperparameter Tuning with Nested Cross-Validation**

   After identifying the best-performing models, I fine-tuned their hyperparameters using nested cross-validation with the function `HT_with_nested_CV`. This method involves two layers of cross-validation: an outer loop to estimate model performance and an inner loop to tune hyperparameters. The outer loop splits the data into training and testing sets multiple times, while the inner loop further splits the training data for hyperparameter tuning.

By combining extensive model comparison with nested cross-validation for hyperparameter tuning, this approach ensures an efficient and unbiased selection of the best predictive algorithm.

- **Inner cross-validation loop preprocessing**: Ensuring that preprocessing steps are applied within the inner cross-validation loop can help avoid potential data leakage.

- **Time series methods**: Implementing time series methods alongside classic machine learning estimators could enhance the model's ability to capture temporal trends and dependencies in the data.
